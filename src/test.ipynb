```python
from __future__ import annotations

from datetime import datetime
from typing import Tuple, Union

from wetterdienst.provider.dwd.observation import DwdObservationRequest

DateLike = Union[str, datetime]


def dwd_daily_met_by_distance(
    latlon: Tuple[float, float],
    distance_km: float,
    start_date: DateLike,
    end_date: DateLike,
    periods: str | None = None,   # e.g., "recent" for recent data; None for full archive range
    drop_nulls: bool = True,
):
    """
    Get DWD stations within `distance_km` of `latlon` and fetch selected daily climate variables.

    Variables fetched:
      - daily/kl: humidity, wind_speed, temperature_air_max_2m, temperature_air_min_2m,
                  temperature_air_mean_2m, precipitation_height
      - daily/solar: radiation_global

    Returns
    -------
    stations_df : polars.DataFrame
        Station metadata for stations within the distance.
    values_df : polars.DataFrame
        Daily time series values for all those station ids and selected variables.
    """

    params = [
        ("daily", "kl", "humidity"),
        ("daily", "kl", "wind_speed"),
        ("daily", "kl", "temperature_air_max_2m"),
        ("daily", "kl", "temperature_air_min_2m"),
        ("daily", "kl", "temperature_air_mean_2m"),
        ("daily", "kl", "precipitation_height"),
        ("daily", "solar", "radiation_global"),
    ]

    # 1) Find stations within distance
    req = DwdObservationRequest(
        parameters=params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )

    stations = req.filter_by_distance(latlon=latlon, distance=distance_km)
    stations_df = stations.df

    # 2) Extract station IDs (robust to minor column naming differences)
    possible_cols = ["station_id", "stationid", "station_id_dwd"]
    station_id_col = next((c for c in possible_cols if c in stations_df.columns), None)
    if station_id_col is None:
        raise KeyError(
            f"Couldn't find station id column. Available columns: {stations_df.columns}"
        )

    station_ids = stations_df.select(station_id_col).to_series().to_list()
    if not station_ids:
        # Return stations (empty) and empty values df
        return stations_df, stations_df.clear()

    # 3) Fetch values for those stations (selected variables only)
    values = (
        req.filter_by_station_id(station_id=tuple(station_ids))
           .values.all()
           .df
    )

    if drop_nulls:
        values = values.drop_nulls()

    return stations_df, values


# --- Example ---
frankfurt = (51.11, 9.68)

stations_df, values_df = dwd_daily_met_by_distance(
    latlon=frankfurt,
    distance_km=30,
    start_date="2018-01-01",
    end_date="2020-01-01",
)


```

```python
from datetime import datetime
from typing import Tuple, Union

from wetterdienst.provider.dwd.observation import DwdObservationRequest

DateLike = Union[str, datetime]


def dwd_daily_met_distance_plus_solar_rank(
    latlon: Tuple[float, float],
    distance_km: float,
    start_date: DateLike,
    end_date: DateLike,
    solar_rank: int = 3,
    periods: str | None = None,
    drop_nulls: bool = True,
):
    """
    KL variables: stations within distance_km
    Solar variable: closest `solar_rank` stations

    Returns
    -------
    stations_df : polars.DataFrame
    values_df   : polars.DataFrame
    """

    kl_params = [
        ("daily", "kl", "humidity"),
        ("daily", "kl", "wind_speed"),
        ("daily", "kl", "temperature_air_max_2m"),
        ("daily", "kl", "temperature_air_min_2m"),
        ("daily", "kl", "temperature_air_mean_2m"),
        ("daily", "kl", "precipitation_height"),
    ]

    solar_params = [
        ("daily", "solar", "radiation_global"),
    ]

    # ---------- KL (distance based) ----------
    kl_req = DwdObservationRequest(
        parameters=kl_params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )

    kl_stations = kl_req.filter_by_distance(latlon=latlon, distance=distance_km)
    kl_df = kl_stations.df

    # ---------- SOLAR (rank based) ----------
    solar_req = DwdObservationRequest(
        parameters=solar_params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )

    solar_stations = solar_req.filter_by_rank(latlon=latlon, rank=solar_rank)
    solar_df = solar_stations.df

    # ---------- Combine station lists ----------
    stations_df = kl_df.vstack(solar_df).unique()

    # station id column (robust)
    possible_cols = ["station_id", "stationid", "station_id_dwd"]
    station_id_col = next((c for c in possible_cols if c in stations_df.columns), None)
    if station_id_col is None:
        raise KeyError(f"No station id column found: {stations_df.columns}")

    station_ids = stations_df.select(station_id_col).to_series().to_list()

    if not station_ids:
        return stations_df, stations_df.clear()

    # ---------- Fetch values (KL + Solar together) ----------
    all_params = kl_params + solar_params

    values_req = DwdObservationRequest(
        parameters=all_params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )

    values_df = (
        values_req
        .filter_by_station_id(station_id=tuple(station_ids))
        .values.all()
        .df
    )

    if drop_nulls:
        values_df = values_df.drop_nulls()

    return stations_df, values_df, kl_df, solar_df


# ---------------- Example ----------------

frankfurt = (51.11, 9.68)

stations_df, values_df, kl_df, solar_df = dwd_daily_met_distance_plus_solar_rank(
    latlon=frankfurt,
    distance_km=30,
    start_date="2018-01-01",
    end_date="2020-01-01",
)

```

```python
from datetime import datetime
from typing import Tuple, Union

from wetterdienst.provider.dwd.observation import DwdObservationRequest

DateLike = Union[str, datetime]


def dwd_daily_met_distance_plus_solar_rank(
    latlon: Tuple[float, float],
    distance_km: float,
    start_date: DateLike,
    end_date: DateLike,
    solar_rank: int = 3,
    periods: str | None = None,
    drop_nulls: bool = True,
):
    """
    KL variables: stations within distance_km
    Solar variable: closest `solar_rank` stations (enforced via .head()).

    Returns
    -------
    stations_df : polars.DataFrame
    values_df   : polars.DataFrame
    """

    kl_params = [
        ("daily", "kl", "humidity"),
        ("daily", "kl", "wind_speed"),
        ("daily", "kl", "temperature_air_max_2m"),
        ("daily", "kl", "temperature_air_min_2m"),
        ("daily", "kl", "temperature_air_mean_2m"),
        ("daily", "kl", "precipitation_height"),
    ]

    solar_params = [
        ("daily", "solar", "radiation_global"),
    ]

    # ---- KL (distance) ----
    kl_req = DwdObservationRequest(
        parameters=kl_params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )
    kl_df = kl_req.filter_by_distance(latlon=latlon, distance=distance_km).df

    # ---- Solar (rank) ----
    solar_req = DwdObservationRequest(
        parameters=solar_params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )

    # filter_by_rank() often sorts but doesn't truncate -> enforce top-N manually
    solar_df = (
        solar_req
        .filter_by_rank(latlon=latlon, rank=solar_rank)
        .df
        .sort("distance")
        .head(solar_rank)
    )

    # ---- Combine stations, unique by station_id ----
    stations_df = kl_df.vstack(solar_df).unique(subset=["station_id"])

    station_ids = stations_df.select("station_id").to_series().to_list()
    if not station_ids:
        return stations_df, stations_df.clear()

    # ---- Fetch values for merged stations (KL + Solar) ----
    all_params = kl_params + solar_params

    values_req = DwdObservationRequest(
        parameters=all_params,
        periods=periods,
        start_date=start_date,
        end_date=end_date,
    )
    values_df = (
        values_req
        .filter_by_station_id(station_id=tuple(station_ids))
        .values.all()
        .df
    )

    if drop_nulls:
        values_df = values_df.drop_nulls()

    return stations_df, values_df, kl_df, solar_df


# ---------------- Example ----------------
frankfurt = (51.11, 9.68)

stations_df, values_df, kl_df, solar_df = dwd_daily_met_distance_plus_solar_rank(
    latlon=frankfurt,
    distance_km=30,
    solar_rank=3,
    start_date="2018-01-01",
    end_date="2020-01-01",
)


```
